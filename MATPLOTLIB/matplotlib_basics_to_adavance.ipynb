```python
#Color(hex) and line(width ans style) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color='green')
plt.plot(batsman['index'],batsman['RG Sharma'],color='red')
plt.title('Rohit Sharma Vs Virat Kohli Carrer Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# Line style
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dashdot',linewidth=4)
plt.title('Rohit Sharma Vs Virat Kohli Carrer Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
#adding legend
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed',linewidth=3,marker='D',markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dashdot',linewidth=2,marker='+',markersize=8,label='Rohit')
plt.title('Rohit Sharma Vs Virat Kohli Carrer Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend(loc='best')
#other loc 
'''
'best', 'upper right', 
'upper left', 'lower left', 
'lower right', 'right', 
'center left', 'center right', 
'lower center', 'upper center', 'center'
'''
```

```python
#Annotations
batter=pd.read_csv('dataset/batter.csv')
df.shape

batter
```

```python
#Annotations
plt.figure(figsize=(15,8))
plt.scatter(batter['avg'],batter['strike_rate'],s=batter['runs'])
for i in range(batter.shape[0]):
    plt.text(batter['avg'].values[i],batter['strike_rate'].values[i],batter['batter'].values[i])
```

```python
#Horizontal and vertical lines
plt.figure(figsize=(15,8))
plt.scatter(batter['avg'],batter['strike_rate'],s=batter['runs'])

plt.axhline(140,color='red')
plt.axhline(130,color='orange')
plt.axvline(30,color='red')

for i in range(batter.shape[0]):
    plt.text(batter['avg'].values[i],batter['strike_rate'].values[i],batter['batter'].values[i],fontdict={'size':10,'color':'red'})
```

```python
#Sub-Plots-A diff way to plot graphs

fig,ax=plt.subplots(figsize=(10,7))
ax.scatter(batter['avg'],batter['strike_rate'])
ax.set_xlabel('Average')
ax.set_ylabel('Strike_rate')
```

```python
#what if you want to plot muliple graphs in o


fig,ax=plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(15,7))
ax[0].scatter(batter['avg'],batter['strike_rate'])
ax[1].scatter(batter['avg'],batter['runs'])

ax[0].set_title('Avg Vs Strike Rate')
ax[0].set_ylabel('Strike Rate')

ax[1].set_title('Avg Vs Runs')
ax[1].set_ylabel('Runs')
```

```python
#making both histogram and scatter in one
#what if you want to plot muliple graphs in o


fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(15,7))
ax[0,0].scatter(batter['avg'],batter['strike_rate'])
ax[0,1].scatter(batter['avg'],batter['runs'])
ax[1,0].hist(batter['strike_rate'])
ax[1,1].hist(batter['runs'])

ax[0,0].set_title('Avg Vs Strike Rate')
ax[0,0].set_ylabel('Strike Rate')

ax[0,1].set_title('Avg Vs Runs')
ax[0,1].set_ylabel('Runs')
```

```python
fig=plt.figure()
ax1=fig.add_subplot(2,2,1)
ax1.scatter(batter['avg'],batter['strike_rate'])

ax2=fig.add_subplot(222)
ax2.scatter(batter['avg'],batter['runs'])

ax3=fig.add_subplot(2,2,3)
ax3.hist(batter['strike_rate'])

ax4=fig.add_subplot(224)
ax4.hist(batter['runs'])
```

```python
#3D surface plot
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
xx,yy=np.meshgrid(x,y)
z=xx**2+yy**2
z.shape
```

```python
fig=plt.figure(figsize=(15,7))
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z=(np.sin(xx))**2+(np.cos(yy))**2
fig=plt.figure(figsize=(15,7))
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```
