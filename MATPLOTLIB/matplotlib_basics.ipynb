```python
# required library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#plotting a simple function
price=[40000,50000,60000,30000,450000]
year=[2020,2021,2022,2023,2024]

plt.plot(year,price)
```

```python
#from pandas dataframe
batsman=pd.read_csv('dataset/sharma-kohli.csv')
batsman

plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
#Lebels ans title 
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('Rohit Sharma Vs Virat Kohli Carrer Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
#Color(hex) and line(width ans style) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color='green')
plt.plot(batsman['index'],batsman['RG Sharma'],color='red')
plt.title('Rohit Sharma Vs Virat Kohli Carrer Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# Line style
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dashdot',linewidth=4)
plt.title('Rohit Sharma Vs Virat Kohli Carrer Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# using marker and marker size
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dashdot',linewidth=2,marker='+',markersize=8)
plt.title('Rohit Sharma Vs Virat Kohli Carrer Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
#adding legend
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed',linewidth=3,marker='D',markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dashdot',linewidth=2,marker='+',markersize=8,label='Rohit')
plt.title('Rohit Sharma Vs Virat Kohli Carrer Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend(loc='best')
#other loc 
'''
'best', 'upper right', 
'upper left', 'lower left', 
'lower right', 'right', 
'center left', 'center right', 
'lower center', 'upper center', 'center'
'''
```

```python
#limiting axes
price=[40000,50000,60000,30000,4500000]
year=[2020,2021,2022,2023,2024]

plt.plot(year,price)
plt.ylim(0,200000)
plt.xlim(2020,2023)
```

```python
#adding grid and exploring show
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed',linewidth=3,marker='D',markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dashdot',linewidth=2,marker='+',markersize=8,label='Rohit')
plt.title('Rohit Sharma Vs Virat Kohli Carrer Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend(loc='best')
plt.grid()
plt.show()
```

```python
df=pd.read_csv('dataset/batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'],df['2015'])
plt.bar(df['batsman'],df['2016'])
plt.bar(df['batsman'],df['2017'])
#so it is overlapping on eachother
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2,color='red')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='blue')
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2,color='orange')
#but number is seeing in x-axis so.. by using xticks we can
plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```

```python
child=[10,5,52,32,]
color=['redredredred','orangeorangeorangeorange','blueblueblueblue','greengreengreengreen']
plt.bar(color,child,color='orange')
plt.xticks(rotation='vertical')
plt.show()
```

```python
#stacked bar chart
plt.bar(df['batsman'],df['2015'],color='red',label='2015')
plt.bar(df['batsman'],df['2016'],bottom=df['2015'],color='blue',label='2016')
plt.bar(df['batsman'],df['2017'],color='yellow',bottom=df['2016']+df['2015'],label='2017')
plt.legend()
plt.show()
```
