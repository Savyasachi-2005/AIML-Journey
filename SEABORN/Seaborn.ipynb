```python
# Seaborn is easy to use
# Better version of matplotlib
# More graphs included compare to matplotlib
# and it is famous

#types of Fuction
'''
Figure level
Axis level
'''
#Main Classfication
'''
Relational Plot - Scatterplot,lineplot
Distribution Plot - Histogram,kdeplot,rugplot 
Categorical Plot - Barplot,Countplot,Boxplot,violenplot,swarmplot
Regression Plot - Regplot
Matric Plot - HeatMap,ClusterMap
Miltiplot - Jointplot,pairplot
'''
```

```python
import seaborn as sns
import matplotlib as plt
import plotly.express as px
```

```python
#Relational Plot 
#Scatter Plot
tips=sns.load_dataset('tips')
```

```python
#Scatter Plot (it is plot mainly in numerical values)
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='time',size='size')
```

```python
#replot -> figure level -> square shape
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',hue='sex',style='time',size='size')
```

```python
#line plot
gap=px.data.gapminder()
temp_gap=gap[gap['country']=='India']
temp_gap
```

```python
#axis level function
sns.lineplot(data=temp_gap,x='year',y='lifeExp')
```

```python
#using relplot
sns.relplot(data=temp_gap,x='year',y='lifeExp',kind='line')
```

```python
#hue,style
temp_gap=gap[gap['country'].isin(['India','Brazil','Germany'])]
temp_gap
```

```python
sns.relplot(data=temp_gap,x='year',y='lifeExp',kind='line',hue='country',style='continent',size='continent')
```

```python
# Facet Plot - plotting multiple graphs on categorical data it will only with relplot not with scatter plot or lineplot 
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',col='sex',size='size',row='day')
```

```python
#col wrap 
sns.relplot(data=gap,x='lifeExp',y='gdpPercap',kind='scatter',col='year',col_wrap=3)
```

```python
#Distribution Data
'''
* used for univariate analysis
* used to find out the distribution
* range of the observation
* central tendency
* is the data bimodel?
* are the outliers?
'''
'''
plots under the DD - histplot,kdeplot,rugplot
'''
```

```python
#Hisplot - Histogram
#figure level -> displot
#axis level - > histplot,kdeplot,rugplot
```

```python
sns.histplot(data=tips,x='total_bill')
```

```python
#using figure level and bin
sns.displot(data=tips,kind='hist',x='total_bill',bins=5)
```

```python
#can be applied on categorical also
sns.displot(data=tips,x='day',kind='hist')
```

```python
#hue parameter, element
sns.displot(data=tips,kind='hist',hue='sex',x='tip',element='step')
```

```python
#using titanic dataset
titanic=sns.load_dataset('titanic')
titanic
```

```python
sns.displot(data=titanic,kind='hist',x='age',element='step',col='sex')

#KDE plot
'''
A Kernel Density Estimate (KDE) plot is a visualization that provides a smooth, 
continuous curve representing the probability density function of a continuous variable. 
It uses a kernel function centered on each data point, summing them to create the overall density curve. 
The bandwidth parameter controls the smoothness of the curve. 
'''
```

```python
sns.kdeplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True)
```

```python
#Rugplot
#plot marginal distribution by drawing ticks along the x and y axes
# this function is intended to complement other plots by showing the location of 
# individual observations in an unobstructive way
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True)
sns.rugplot(data=tips,x='total_bill',hue='sex')
```

```python
#we can make bivariate histogram 
# A bivariate histogram bind the data within rectangles that title the plot
# and then shows the count of observation within each rectangle with the fill color
sns.histplot(data=tips,x='total_bill',y='tip')
```

```python
sns.displot(data=tips,x='total_bill',y='tip',kind='kde')
```

```python
#Matrix
#HeatMap
#Plot rectangular data as a color-encoded matrix
gap
```

```python
temp_df=gap.pivot(index='country',columns='year',values='lifeExp')
#axes leve function
#plt.figure(figsize=(15,10))
sns.heatmap(temp_df)
```

```python
#annotation linewidth cmap
temp_df=gap[gap['continent']=='Asia'].pivot(index='country',columns='year',values='lifeExp')
sns.heatmap(temp_df,annot=True,linewidth=0.5,cmap='summer')
```
